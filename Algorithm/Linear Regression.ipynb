```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#生成线性训练数据
X = np.arange(0, 1, 0.01)
X = np.c_[np.ones(X.shape[0]), X]

# y = x + 1
y = []
for i in X:
    y.append(i[1] + 1 + np.random.normal(0,0.1)) 
y = np.array(y)
```

```python
#Linear Rregression closed-formed solution
def Lin_reg(X, y):
    X = np.mat(X)
    y = np.mat(y).T
    w = (X.T * X).I * X.T * y
    return np.array(w).flatten()
```

```python
w = Lin_reg(X, y)
plt.scatter(X[:,1],y)
x = np.arange(0, 1, 0.01)
w_g = w[0] + w[1]*x
w_f = x + 1
plt.plot(x, w_g, label='y={:.2f}x+{:.1f}'.format(w[1],w[0]))
plt.plot(x, w_f, label='f')
plt.legend()
```
